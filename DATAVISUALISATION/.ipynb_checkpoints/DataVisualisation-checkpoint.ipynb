```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
years = [2011, 2015, 2019, 2023, 2011, 2015, 2019, 2023]
country = ['INDIA', 'INDIA', 'INDIA', 'INDIA', 'PAK', 'PAK', 'PAK', 'PAK']
GDP = [1, 5, 6, 15, 2, 2, 3, 1]

df_gdp = pd.DataFrame({'years': years, 'country': country, 'GDP': GDP})

# Map countries to numerical values for coloring
color_map = {'INDIA': 1, 'PAK': 0}
df_gdp['country_numeric'] = df_gdp['country'].map(color_map)

# Plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    df_gdp['years'],
    df_gdp['GDP'],
    c=df_gdp['country_numeric'],
    cmap='jet',
    s=100,  # size of points
    edgecolors='k'
)

# Annotate points with country names for clarity
for i, txt in enumerate(df_gdp['country']):
    plt.annotate(txt, (df_gdp['years'][i] + 0.2, df_gdp['GDP'][i]))

plt.title("Years vs GDP Growth")
plt.xlabel('Years')
plt.ylabel('GDP')
plt.grid(True)
plt.colorbar(scatter, label='Country (1=India, 0=Pak)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame({"Batsmen":["AB","Dhoni","Sachin","VK"],"2015": [2000,330,550,2222],"2016": [400,430,850,222],"2017": [900,930,950,929]})
```

```python
# First series
plt.bar(df['Batsmen'], df['2015'], label='2015')

# Second series stacked on top of 2015
plt.bar(df['Batsmen'], df['2016'], bottom=df['2015'], label='2016')

# Third series stacked on top of 2015 + 2016
plt.bar(df['Batsmen'], df['2017'], bottom=df['2015'] + df['2016'], label='2017')

plt.xlabel('Batsmen')
plt.ylabel('Runs')
plt.title('Runs scored by Batsmen (20152017)')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
fig,ax =  plt.subplots(nrows=2,ncols=1,figsize=(18, 12))
years=[2011,2015,2019,2023]
gdp_of_india = [1,5,6,15]
gdp_of_pakistan = [1,2,3,1]
ax[0].scatter(years,gdp_of_pakistan)
ax[0].set_title("GDP GROWTH OF PAKISTAN")
ax[0].set_xlabel("X-Axis")
ax[0].set_ylabel("Y-Axis")
fig.show()
ax[1].scatter(years,gdp_of_india)
ax[1].set_title("GDP GROWTH OF INDIA")
ax[1].set_xlabel("X-Axis")
ax[1].set_ylabel("Y-Axis")
fig.show()
```

```python
fig =  plt.figure(figsize=(18,10))
ax=fig.add_subplot(1,3,1)
years=[2011,2015,2019,2023]
gdp_of_india = [1,5,6,15]
gdp_of_pakistan = [1,2,3,1]
gdp_of_bangladesh = [1,2,6,3]

ax.scatter(years,gdp_of_pakistan)
ax.set_title("GDP GROWTH OF PAKISTAN")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")

ax1=fig.add_subplot(1,3,2)
ax1.scatter(years,gdp_of_india)
ax1.set_title("GDP GROWTH OF INDIA")
ax1.set_xlabel("X-Axis")
ax1.set_ylabel("Y-Axis")
fig.show()

ax2=fig.add_subplot(1,3,3)
ax2.scatter(years,gdp_of_bangladesh)
ax2.set_title("GDP GROWTH OF BANGLADESH")
ax2.set_xlabel("X-Axis")
ax2.set_ylabel("Y-Axis")
fig.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated runs per ball for 5 overs (5 overs  6 balls = 30 balls)
# Each row = one over, each column = one ball in that over
runs = np.array([
    [1, 0, 4, 6, 1, 1],
    [0, 2, 1, 0, 4, 1],
    [6, 1, 0, 1, 2, 0],
    [1, 4, 1, 1, 0, 6],
    [0, 1, 0, 2, 4, 1]
])

# Plotting the heatmap
plt.figure(figsize=(8, 5))
heatmap = plt.imshow(runs, cmap='YlOrRd', interpolation='nearest')

# Add colorbar
plt.colorbar(heatmap, label='Runs Scored')

# Label axes
plt.title("IPL Runs Per Ball (5 Overs)")
plt.xlabel("Ball Number in Over")
plt.ylabel("Over Number")

# Tick labels
plt.xticks(np.arange(6), labels=[f'Ball {i+1}' for i in range(6)])
plt.yticks(np.arange(5), labels=[f'Over {i+1}' for i in range(5)])

plt.tight_layout()
plt.show()
```

```python
#Parent level plots
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='sex',style='time')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='line',hue='sex',style='time')
```

```python
#A facet plot (also called faceting or small multiples) is a way to split data across multiple subplots based on a category. Each subplot shows a subset of the data  usually filtered by a specific group  but all share the same axes or structure, making comparisons easier.



sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',col='sex',row='day')
```

```python
# scatter plot -> axes level function child level
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',style='time',size='size')
```
