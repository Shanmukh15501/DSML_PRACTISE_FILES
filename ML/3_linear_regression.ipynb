# Cost Function
- A cost function measures how wrong a machine learning model's predictions are. It's used to optimize the model by minimizing the "error" between predictions and actual values. Think of it as a penalty for bad predictions.

- Mean Absolute Error calculates the average difference between the calculated values and actual values. It is also known as scale-dependent accuracy as it calculates error in observations taken on the same scale used to predict the accuracy of the machine learning model.



- Mean Squared Error (MSE): This is a commonly used cost function for regression problems. It calculates the average squared difference between the predicted and actual values.

- 



````python
import numpy as np

y_pred = np.array([1.2, 2.1, 3.5, 4.8])
y_actual = np.array([1, 2, 3, 5])
n = len(y_pred)

mse = 1/n * np.sum((y_pred - y_actual)**2)
print('MSE:', mse)

````

- Binary Cross-Entropy/Logloss: This is a commonly used cost function for binary classification problems. It measures the difference between the predicted and actual probabilities of the binary output variable.





````python
import numpy as np

y_pred = np.array([0.2, 0.8, 0.6, 0.3])
y_actual = np.array([0, 1, 1, 0])
n = len(y_pred)

bce = -1/n * np.sum(y_actual * np.log(y_pred) + (1 - y_actual) * np.log(1 - y_pred))
print('Binary Cross-Entropy:', bce)

````

- Categorical Cross-Entropy: This is a commonly used cost function for multi-class classification problems. It measures the difference between the predicted and actual probabilities of the multiple output classes.



````python
import numpy as np

y_pred = np.array([[0.1, 0.3, 0.6], [0.2, 0.7, 0.1], [0.9, 0.05, 0.05]])
y_actual = np.array([[0, 0, 1], [0, 1, 0], [1, 0, 0]])
n = len(y_pred)

cce = -1/n * np.sum(np.sum(y_actual * np.log(y_pred)))
print('Categorical Cross-Entropy:', cce)
````

- Hinge Loss: This is a cost function commonly used for training models for binary classification tasks using Support Vector Machines (SVMs). It penalizes misclassified samples and aims to maximize the margin between the decision boundary and the training samples.
- HL = max(0, 1  y_actual * y_pred)



````python
import numpy as np

y_pred = np.array([0.2, 0.8, 0.6, 0.3])
y_actual = np.array([-1, 1, 1, -1])
n = len(y_pred)

hl = np.mean(np.maximum(0, 1 - y_actual * y_pred))
print('Hinge Loss:', hl)
````

-  Kullback-Leibler Divergence: This cost function measures the difference between the predicted and actual probability distributions and is commonly used in tasks such as generative modeling.
- KL = sum(y_actual * log(y_actual / y_pred))





````python
import numpy as np

y_pred = np.array([0.1, 0.5, 0.4])
y_actual = np.array([0.2, 0.3, 0.5])

kl = np.sum(y_actual * np.log(y_actual / y_pred))
print('Kullback-Leibler Divergence:', kl)
````

````python
#https://en.wikipedia.org/wiki/Feature_scaling
def gradient_descent(x_scaled, y_scaled, lr=0.1, epochs=1000):
    # Scale x and y using Min-Max Scaling
    x_min, x_max = x.min(), x.max()
    y_min, y_max = y.min(), y.max()

    x_scaled = (x - x_min) / (x_max - x_min)
    y_scaled = (y - y_min) / (y_max - y_min)

    # Initialize parameters
    b = 0.0  # Intercept
    m = 0.0  # Slope
    n = len(y_scaled)  # Number of data points

    # Perform gradient descent
    for epoch in range(epochs):
        y_pred = b + m * x_scaled  # Predicted y values
        error = y_scaled - y_pred  # Error in prediction
        cost = np.mean(error ** 2)   # Mean squared error

        # Calculate gradients
        db = -2 * np.mean(error)  # Derivative w.r.t. intercept b
        dm = -2 * np.mean(error * x_scaled)  # Derivative w.r.t. slope m

        # Update parameters
        b -= lr * db
        m -= lr * dm

        # Optional: Print cost every 100 iterations to monitor progress
        if epoch % 100 == 0:
            print(f"Epoch {epoch}: Cost = {cost}, b = {b}, m = {m}")

    # Scale back the coefficients to original scale
    b_original = b * (y_max - y_min) + y_min - m * (y_max - y_min) * x_min / (x_max - x_min)
    m_original = m * (y_max - y_min) / (x_max - x_min)

    return b_original, m_original


if __name__ == "__main__":
    df = pd.read_csv("D:\\utils\\DataSets\\home_single_prices.csv")
    x = df["area_sqr_ft"].to_numpy()
    y = df["price_lakhs"].to_numpy()

    b, m = gradient_descent(x, y)

    print(f"Final Results: m={m}, b={b}")



````

The preference for Mean Squared Error (MSE) over Mean Absolute Error (MAE) in gradient descent largely boils down to the mathematical properties of these loss functions, particularly their differentiability. Here's a breakdown:

**1. Differentiability:**

* **MSE:**
    * The MSE function is differentiable at all points. This is crucial for gradient descent, which relies on calculating the gradient (the derivative) of the loss function to determine the direction of parameter updates.
    * The derivative of MSE is a linear function of the error, which makes it well-behaved and facilitates stable convergence of gradient descent.
* **MAE:**
    * The MAE function is not differentiable at zero. This can cause problems for gradient descent, as the algorithm may struggle to find the optimal solution when errors are close to zero.
    * While there are ways to work around this, the non-differentiability of MAE can make optimization more complex.

**2. Smoothness:**

* **MSE:**
    * MSE produces a smooth, parabolic curve. This smoothness helps gradient descent to converge steadily towards the minimum.
* **MAE:**
    * MAE produces a V-shaped curve, which has a sharp corner at zero. This can lead to oscillations and slower convergence of gradient descent.

**3. Behavior with Outliers:**

* **MSE:**
    * MSE penalizes larger errors more heavily due to the squaring operation. This can be beneficial when large errors are particularly undesirable.
* **MAE:**
    * MAE treats all errors equally, making it more robust to outliers. However, this can also mean that gradient descent may not be as effective in minimizing large errors.

**In summary:**

The smooth, differentiable nature of MSE makes it more compatible with gradient descent, leading to more stable and efficient optimization.

While MAE has its advantages, particularly in robustness to outliers, its non-differentiability can pose challenges for gradient-based optimization.

Therefore, for the mathematical reasons related to the functionality of gradient descent, MSE is very often the preferred loss function.

## Regression Evaluation Metrics

Here's a breakdown of common evaluation metrics used in regression, along with their formulas:

**1. Mean Absolute Error (MAE)**

* **Description:** The average of the absolute differences between predicted and actual values.
* **Formula:**
    ```
    MAE = (1/n) *  |y_i - _i|
    ```
    * Where:
        * `n` is the number of data points.
        * `y_i` is the actual value.
        * `_i` is the predicted value.

**2. Mean Squared Error (MSE)**

* **Description:** The average of the squared differences between predicted and actual values.
* **Formula:**
    ```
    MSE = (1/n) *  (y_i - _i)^2
    ```
    * Where:
        * `n` is the number of data points.
        * `y_i` is the actual value.
        * `_i` is the predicted value.

**3. Root Mean Squared Error (RMSE)**

* **Description:** The square root of the MSE. It provides an error metric in the same units as the target variable.
* **Formula:**
    ```
    RMSE = MSE = ( (1/n) *  (y_i - _i)^2 )
    ```

**4. R-squared (Coefficient of Determination)**

* **Description:** Measures the proportion of the variance in the dependent variable that is predictable from the independent variable(s). It indicates how well the model fits the data.
* **Formula:**
    ```
    R = 1 - (SS_res / SS_tot)
    ```
    * Where:
        * `SS_res` is the sum of squared residuals (errors): ` (y_i - _i)^2`.
        * `SS_tot` is the total sum of squares: ` (y_i - mean(y))^2`.

**5. Adjusted R-squared**

* **Description:** A modified version of R-squared that adjusts for the number of predictors in the model. It penalizes the inclusion of unnecessary predictors.
* **Formula:**
    ```
    Adjusted R = 1 - [ (1 - R) * (n - 1) / (n - k - 1) ]
    ```
    * Where:
        * `n` is the number of data points.
        * `k` is the number of predictors.

**6. Mean Absolute Percentage Error (MAPE)**

* **Description:** The average of the absolute percentage differences between predicted and actual values.
* **Formula:**
    ```
    MAPE = (1/n) *  |(y_i - _i) / y_i| * 100%
    ```
    * Where:
        * `n` is the number of data points.
        * `y_i` is the actual value.
        * `_i` is the predicted value.
    * **Note:** MAPE is undefined when any `y_i` is zero.

**7. Median Absolute Error (MedAE)**

* **Description:** The median of the absolute differences between predicted and actual values. Less sensitive to outliers than MAE.
* **Formula:**
    ```
    MedAE = median(|y_i - _i|)
    ```
    * Where:
        * `y_i` is the actual value.
        * `_i` is the predicted value.

**Key Considerations:**

* The choice of metric depends on the specific problem and the importance of different types of errors.
* MSE and RMSE penalize larger errors more heavily.
* MAE and MedAE are more robust to outliers.
* R-squared and Adjusted R-squared indicate how well the model fits the data.
* MAPE provides a percentage error, which can be easier to interpret.
