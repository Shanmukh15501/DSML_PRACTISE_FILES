### `Q-1:` Using `groupby` make groups using the `"Pclass"` column and find out the average age and total number of missing values in the `"Age"` column for every group.

````python
# code here

import pandas as pd

url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQjh5HzZ1N0SU7ME9ZQRzeVTaXaGsV97rU8R7eAcg53k27GTstJp9cRUOfr55go1GRRvTz1NwvyOnuh/pub?gid=1562145139&single=true&output=csv"
titanic_df = pd.read_csv(url)
titanic_df.head(1)
````

````python
titanic_df_grp = titanic_df.groupby('Pclass')
print('Total Dataset Shape',titanic_df.shape)
print('Total null values in Age',titanic_df[titanic_df['Age'].isnull()].shape)

result = titanic_df_grp.agg(
    Age_mean=('Age', 'mean'),
    Age_missing_count=('Age', lambda x: x.isnull().sum())
)
result
````

### `Q-2:` Using `groupby` make groups using the `"Pclass"` column and fill every group's `"Embarked"` column's missing values with the mode value of that group. After that, print every group's `"Embarked"` column's value counts in ascending order.

````python
titanic_df_grp = titanic_df.groupby('Pclass')

#Another way to find null values count
print("Before Update Count",titanic_df_grp.aggregate(Total=("Embarked","count")))

def update_null_values(value,data):
    if pd.isnull(value):
        return data['Embarked'].mode()[0]
    return value   
#Another way to find null values count
#titanic_df_grp['Embarked'].apply(lambda x: x.isnull().sum())




for group,data in titanic_df_grp:
    #why we need this step because group by while updating this data in the loop
    #it actually sends a copy does not update original 
    updated_embarked = data['Embarked'].apply(lambda x: update_null_values(x, data))

    titanic_df.loc[titanic_df['Pclass']==group,'Embarked']=updated_embarked

  
print("After Update Count",titanic_df_grp.aggregate(Total=("Embarked","count")))

````

````python
#Method-2
titanic_df['Embarked'] = titanic_df.groupby('Pclass')['Embarked'].transform(lambda x: x.fillna(x.mode()[0]))
print("After Update Count",titanic_df.groupby('Pclass').aggregate(Total=("Embarked","count")))
````

### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
# code here
import pandas as pd

fifa_df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vT3D_x_4DS6d51LKJ7ze1sxT5WpV5uiSVOFYHLwBiGru6vFyVv5h5-83AwFjxWYiWfCDjDAaarHAV-k/pub?gid=0&single=true&output=csv")
fifa_df.head(1)
````

````python
import numpy as np

fifa_df['ZScore_Passes']=[np.NaN]*fifa_df.shape[0]
fifa_df['ZScore_Passes_Completed']=[np.NaN]*fifa_df.shape[0]
fifa_df['ZScore_Attempted_Line_Breaks']=[np.NaN]*fifa_df.shape[0]
fifa_df['ZScore_Completed_Line_Breaks']=[np.NaN]*fifa_df.shape[0]

#fifa_df[['Attempted Line Breaks', 'Completed Line Breaks', 'Passes', 'Passes Completed']][:2]

fifa_df_grp=fifa_df.groupby('Team')


def applyzscore(data,group):
    Z=(data - np.mean(data))/np.std(data)
    return Z
for group,data in fifa_df_grp:
    fifa_df.loc[fifa_df['Team']==group,'ZScore_Passes']=data.apply(lambda x:applyzscore(data['Passes'],group))
    fifa_df.loc[fifa_df['Team']==group,'ZScore_Passes_Completed']=data.apply(lambda x:applyzscore(data['Passes Completed'],group))
    fifa_df.loc[fifa_df['Team']==group,'ZScore_Attempted_Line_Breaks']=data.apply(lambda x:applyzscore(data['Attempted Line Breaks'],group))
    fifa_df.loc[fifa_df['Team']==group,'ZScore_Completed_Line_Breaks']=data.apply(lambda x:applyzscore(data['Completed Line Breaks'],group))


fifa_df_grp.agg({
    "Yellow Cards":["min","max","mean"],
    "Passes":["min","max"],
    "Attempted Line Breaks":["min","max","std"],
    "Possession (%)":"mean"
    
})
    
````

###`Q-7`: Most successful batting pair in IPL. Batting pair who have scored most runs playing together.


````python
# code here
def check_data(x):
    
    return x['batter']+'-'+x['non-striker']

df['Batting_Pair'] = df.apply(lambda x:check_data(x),axis=1)

df['Batting_Pair']=df.apply(lambda x: '-'.join(sorted(x['Batting_Pair'].split('-'))), axis=1)

df3=(df.groupby(['Batting_Pair']).aggregate({'total_run':'sum'}))
df3.sort_values(by='total_run',ascending=False)

````

###`Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
df4 = (
    df.groupby(['Batting_Pair'])
    .agg(
        Total=('total_run', 'sum'),
        Total_Balls=('ballnumber', 'count')
    )
    .assign(
        AVG=lambda x: x['Total'] / df.groupby('Batting_Pair')['isWicketDelivery'].sum()  # Calculate AVG after Total is available
    )
    .sort_values('Total', ascending=False)
    .reset_index()
)

df4['Strike_Rate']=df4['Total']/df4['Total_Balls']*100
df4.head(4)
````
