# Types of Uniform Distribution

###### Discrete Uniform Distribution:
- This type of distribution deals with a finite number of possible outcomes, where each outcome has an equal probability of occurring.   
- A classic example is rolling a fair six-sided die. Each face (1, 2, 3, 4, 5, or 6) has a 1/6 probability of landing face up.   
- In essence, you're dealing with countable outcomes.   

###### Continuous Uniform Distribution:
- This type of distribution deals with an infinite number of possible outcomes within a specific range, where every value within that range has an equal probability.   
- An idealized random number generator producing values between 0 and 1 is a good example. Any value within that interval is equally likely.   
- Here, you're dealing with outcomes that can take on any value within a continuous interval.

# The Pareto distribution
- The Pareto distribution, also known as the 80/20 rule, is a power-law probability distribution that describes situations where a large portion of effects come from a small portion of causes, often used in economics and business to model wealth distribution and other phenomena. 
- This is based on power law, Used for describing Financial data
- https://en.wikipedia.org/wiki/Pareto_distribution
- This distibution tells that 20% plays a vital rolel in remaining 80% of data.

##### The Pareto distribution models situations where a few items contribute to a large portion of the total. Common uses include:

- Economics: Wealth/income inequality.
- Business: Customer/sales analysis (80/20 rule).
- Quality: Identifying major problem causes.
- Risk: Modeling large losses.
- Internet: Traffic patterns.

## Pareto Distribution PDF

The probability density function (PDF) of the Pareto distribution is given by:

$$ f(x; x_m, \alpha) = \begin{cases} \frac{\alpha x_m^\alpha}{x^{\alpha+1}} & \text{if } x \ge x_m \\ 0 & \text{if } x < x_m \end{cases} $$

Where:

* $x_m$ is the minimum possible value of $x$ (scale parameter).
* $\alpha$ is the shape parameter, determining the heaviness of the tail.
* $x$ is the random variable.

**Key Points:**

* The Pareto distribution is defined for $x \ge x_m$.
* It is a power-law distribution, meaning the probability of large values decreases polynomially.
* The shape parameter $\alpha$ controls how quickly the tail decreases. A smaller $\alpha$ indicates a heavier tail (more extreme values).

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the Pareto distribution
alpha = 3
xm = 1

# Create an array of x values
x = np.linspace(0.1, 10, 1000)

# Calculate the y values of the Pareto distribution
y = alpha * (xm**alpha) / (x**(alpha+1))


plt.plot(x,y)
````

## Mathematical Transformers for Data Preprocessing

These transformations are used to modify the distribution of features in your data, often to make them more suitable for machine learning models.

### Types of Mathematical Transformers

* **Function Transformer**:

    * Applies an arbitrary function to your data.
    * You define the function (e.g., `np.log`, `np.exp`, a custom function).
    * Example:

        ```python
        from sklearn.preprocessing import FunctionTransformer
        import numpy as np

        transformer = FunctionTransformer(np.log)
        data_transformed = transformer.fit_transform(data)
        ```

* **Power Transformer**:

    * A family of transformations that aim to make data more Gaussian-like (normal distribution).
    * Includes Box-Cox and Yeo-Johnson transformations.

* **Quantile Transformer**:

    * Transforms the data to follow a uniform or a normal distribution based on the quantiles of the data.
    * Non-linear transformation.
    * Robust to outliers.
    * Example:
        ```python
        from sklearn.preprocessing import QuantileTransformer
        import numpy as np

        transformer = QuantileTransformer(output_distribution='normal') #or 'uniform'
        data = np.array([[1, 2], [3, 4], [5, 6]])
        transformed_data = transformer.fit_transform(data)
        print(transformed_data)
        ```

* **Log Transform**:

    * Applies the logarithm function (typically natural logarithm) to the data.
    * Useful for reducing the skewness of right-skewed data.
    * Helpful for data with a large range of values.
    * Example:

        ```python
        import numpy as np

        data_log_transformed = np.log(data)
        ```

* **Reciprocal Transform**:

    * Applies the function 1/x to the data.
    * Can help with right-skewed data, but less common than log transform.
    * Example:
        ```python
        import numpy as np

        data_reciprocal_transformed = 1 / data
        ```

* **Power Transform**:

    * A general term for transformations that raise the data to a power.
    * Both Box-Cox and Yeo-Johnson are power transforms.

* **Box-Cox**:

    * A power transform that finds the optimal power to make the data as close to a normal distribution as possible.
    * Requires the data to be strictly positive.
    * Example:

        ```python
        from scipy.stats import boxcox
        import numpy as np

        data = np.array([1, 2, 3, 4, 5])  # Example data
        data_transformed, lambda_value = boxcox(data)
        print(data_transformed)
        print(lambda_value)
        ```

* **Yeo-Johnson**:

    * An extension of the Box-Cox transformation that can handle both positive and non-positive data (including zeros and negative values).
    * Example:
        ```python
        from scipy.stats import yeojohnson
        import numpy as np

        data = np.array([-2, -1, 0, 1, 2])
        transformed_data, lambda_value = yeojohnson(data)
        print(transformed_data)
        print(lambda_value)
        ```

### Why Use These Transformations?

These transformations are often used to improve the performance of machine learning models by making the data more suitable for the model's assumptions, such as:

* **Normality**: Many statistical models assume that the data is normally distributed.
* **Linearity**: Some models assume a linear relationship between variables.
* **Equal Variance (Homoscedasticity)**:  Models often assume the variance of the errors is constant.

- Shape: It has a characteristic bell-shaped curve.   

- Mean, Median, and Mode: In a normal distribution, the mean (μ), median, and mode are all equal and located at the center of the distribution.   

- Symmetry: The distribution is perfectly symmetric around the mean.

- Approximately 95% of the data falls within 2 standard deviations of the mean.   
- Approximately 99.7% of the data falls within 3 standard deviations of the mean.

# 1. Cumulative Distribution Function (CDF)

##### Definition:
- The CDF of a normal distribution (or any probability distribution) gives the probability that a random variable will take on a value less than    or equal to a given value.   
- In simpler terms, it tells you the area under the normal curve to the left of a specific point.
- Mathematically, for a random variable X, the CDF, denoted as F(x), is:
- F(x) = P(X ≤ x) 


# 2. Percent Point Function (PPF)

##### Definition:
- The PPF is the inverse of the CDF.   
- It gives you the value of the random variable for which the probability of being less than or equal to that value is a given probability.
- In other words, it tells you the value that corresponds to a specific percentile.

# Standard normal distribution
- A normal distribution with a mean 0 and standard deviation of 1 is called the standard normal distribution
- The key to converting a normal distribution to a standard normal distribution is to calculate z-scores.
 A z-score represents how many standard deviations a data point is away from the mean.   

- Here's the formula:
- z = (X - μ) / σ   

##### Where:
- z is the z-score.   
- X is the data point.   
- μ is the mean of the original distribution.
- σ is the standard deviation of the original distribution.

##### What the Z-Score Tells You

- A positive z-score indicates that the data point is above the mean.   
- A negative z-score indicates that the data point is below the mean.   
- A z-score of 0 indicates that the data point is equal to the mean.   

#### Why Convert to Standard Normal?

- Standardization: It allows you to standardize data from different normal distributions, making it easier to compare and analyze them.   
- Probability Calculations: Standard normal distribution tables (z-tables) provide probabilities associated with different z-scores, enabling you to calculate the likelihood of certain outcomes.   



````python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import normal
from scipy.stats import norm

# Generate a sample from a normal distribution
sample = normal(loc=50, scale=5, size=1000)

# Calculate sample mean and sample standard deviation
sample_mean = sample.mean()
sample_std = sample.std()



# Plot the histogram of the sample
sns.kdeplot(sample, color='skyblue', label='Standard Normal Distribution')
````

# Sampling bias
It means your sample isn't a good reflection of the whole group you're studying. It leads to skewed results because some people are more likely to be included than others. Think of it like only asking tall people about average height – you'll get a wrong answer.

- Selection Bias: Choosing samples in a non-random way.
- Self-Selection Bias: Participants choose to be in the study.
- Convenience Bias: Using easily available participants.
- Survivorship Bias: Focusing on those that "survived" a process.
- Undercoverage Bias: Some groups are poorly represented.

# Standard Error
- Standard Error (SE) is a measure of the statistical accuracy of an estimate. It quantifies how much the sample mean (or other statistic) is likely to vary from the population mean.
# Key Points:
- A smaller standard error indicates a more precise estimate.
- It's closely related to the standard deviation, but they are not the same.
- Standard error is used in confidence intervals and hypothesis testing.
- It's calculated as: SEM = standard deviation / square root of sample size.
- This means that as your sample size increases, the standard error decreases.






````python
import numpy as np

def standard_error(data):
  """Calculates standard error of the mean."""
  std_dev = np.std(data, ddof=1) # Sample standard deviation
  return std_dev / np.sqrt(len(data))

# Example:
data = [1, 2, 3, 4, 5]
se = standard_error(data)
print(f"Standard Error: {se}")
````
