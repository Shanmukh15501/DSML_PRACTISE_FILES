## Normal Distribution (Gaussian Distribution)

- A normal distribution (also called a Gaussian distribution) is a continuous probability distribution that is symmetric about the mean. It's one of the most important probability distributions in statistics, as many real-world phenomena approximate this pattern.   

## Properties of the Normal Distribution:

- Shape: It has a characteristic bell-shaped curve.   

- Mean, Median, and Mode: In a normal distribution, the mean (μ), median, and mode are all equal and located at the center of the distribution.   

- Symmetry: The distribution is perfectly symmetric around the mean.   

- 68-95-99.7 Rule (Empirical Rule):

- Approximately 68% of the data falls within 1 standard deviation (σ) of the mean (μ).   
- Approximately 95% of the data falls within 2 standard deviations of the mean.   
- Approximately 99.7% of the data falls within 3 standard deviations of the mean.

# ZScore

- In statistics, a z-score (also called a standard score) is a way to describe how far a particular data point is from the mean (average) of a dataset. Specifically, it tells you:
- How many standard deviations a data point is away from the mean.   
- Whether the data point is above or below the mean.

### What it does:
- It standardizes data, allowing you to compare values from different datasets.   
- It helps identify outliers (values that are unusually far from the mean).   
- It's used in hypothesis testing and other statistical analyses

### Key points:
- A positive z-score indicates the data point is above the mean.   
- A negative z-score indicates the data point is below the mean.   
- A z-score of 0 means the data point is exactly equal to the mean.

 


```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrames (replace with your actual data)
data1 = {'Virat_Scores': [75, 82, 90, 68, 88, 55, 95, 70, 80, 192]}
data2 = {'Babar_Scores': [60, 78, 85, 72, 90, 65, 88, 75, 80, 95]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Calculate z-scores for Virat's scores
df1['Virat_Zscore'] = stats.zscore(df1['Virat_Scores'])

# Calculate z-scores for Babar's scores
df2['Babar_Zscore'] = stats.zscore(df2['Babar_Scores'])

# Display the DataFrames with z-scores
print("Virat's Scores with Z-scores:\n", df1)
print("\nBabar's Scores with Z-scores:\n", df2)


# Plotting Virat's Scores and Z-scores
plt.figure(figsize=(12, 6))


# Subplot 2: Virat's Z-scores
plt.subplot(1, 2, 1)  # 1 row, 2 columns, second subplot

sns.histplot(df1['Virat_Zscore'], kde=True)
plt.title("Virat's Z-scores Distribution")
plt.xlabel("Z-scores")
plt.ylabel("Frequency")


plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot

sns.histplot(df2['Babar_Zscore'], kde=True)
plt.title("Babar's Z-scores Distribution")
plt.xlabel("Z-scores")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

```

# Sampling bias
It means your sample isn't a good reflection of the whole group you're studying. It leads to skewed results because some people are more likely to be included than others. Think of it like only asking tall people about average height – you'll get a wrong answer.

- Selection Bias: Choosing samples in a non-random way.
- Self-Selection Bias: Participants choose to be in the study.
- Convenience Bias: Using easily available participants.
- Survivorship Bias: Focusing on those that "survived" a process.
- Undercoverage Bias: Some groups are poorly represented.

# Central Limit Theorem (CLT)
- The central limit theorem (CLT) states that the distribution of sample means approximates a normal distribution as the sample size gets larger, regardless of the population's distribution.
- A sufficiently large sample size can predict the characteristics of a population more accurately.
- Sample sizes equal to or greater than 30 are often considered sufficient for the CLT to hold.
- A key aspect of CLT is that the average of the sample means and standard deviations will equal the population mean and standard deviation.
- CLT is useful in finance and investing when analyzing a large collection of securities to estimate portfolio distributions and traits for returns, risk, and correlation.

```python
import numpy as np
import matplotlib.pyplot as plt

#Use np.random.uniform for continuous data (decimals, like 3.14) within a range. 
# Use np.random.randint for discrete data (whole numbers, like 1, 2, 3) within a range.
#uniform: Simulates things like measurements (height, weight).
#randint: Simulates things like dice rolls, counts.population = np.random.uniform(0, 10, 1000)

# Number of samples and sample size
num_samples = 1000
sample_size = 30  # Increase for better CLT effect

# List to store sample means
sample_means = []

population = np.random.uniform(0,100,1000)

# Take multiple samples and calculate their means
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Plot the distribution of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')
plt.title("Distribution of Sample Means")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.show()

#Plot the original distribution.
plt.hist(population, bins=30, density=True, alpha=0.6, color = 'b')
plt.title("Original population Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

# Standard Error
- Standard Error (SE) is a measure of the statistical accuracy of an estimate. It quantifies how much the sample mean (or other statistic) is likely to vary from the population mean.
# Key Points:
- A smaller standard error indicates a more precise estimate.
- It's closely related to the standard deviation, but they are not the same.
- Standard error is used in confidence intervals and hypothesis testing.
- It's calculated as: SEM = standard deviation / square root of sample size.
- This means that as your sample size increases, the standard error decreases.






```python
import numpy as np

def standard_error(data):
  """Calculates standard error of the mean."""
  std_dev = np.std(data, ddof=1) # Sample standard deviation
  return std_dev / np.sqrt(len(data))

# Example:
data = [1, 2, 3, 4, 5]
se = standard_error(data)
print(f"Standard Error: {se}")
```

# Confidence intervals 
- Confidence intervals (CIs) are a way to express the uncertainty around a statistical estimate. They provide a range of values within which we are reasonably confident the true population parameter lies.

- They give you a range of plausible values for a population parameter (like the mean).
- They tell you how confident you can be that the true value is within that range.
### Key components:
- Estimate: The sample statistic (e.g., sample mean).
- Margin of error: The range added and subtracted from the estimate.
- Confidence level: The probability that the interval contains the true population parameter (e.g., 95%).
