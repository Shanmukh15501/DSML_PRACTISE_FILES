# Uniform Distribution:

- A uniform distribution is a probability distribution where all outcomes are equally likely. Imagine rolling a fair die; each face (1 to 6) has an equal chance of appearing. It's like a flat line on a probability graph.
- Problem Statement:

Problem: Simulate rolling a fair six-sided die 1000 times and visualize the distribution of the results.

```python
import random
import matplotlib.pyplot as plt

# Simulate 1000 rolls of a 6-sided die (values between 1 and 6)
rolls = [i for i in range(1,7)]

# Plotting the frequency of each die face
plt.hist(rolls, bins=range(1, 8), edgecolor='black', align='left', rwidth=0.8)
plt.title('Uniform Distribution')
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.xticks(range(1, 7))
plt.show()
```

## Normal Distribution (Gaussian Distribution)

- A normal distribution (also called a Gaussian distribution) is a continuous probability distribution that is symmetric about the mean. It's one of the most important probability distributions in statistics, as many real-world phenomena approximate this pattern.   

## Properties of the Normal Distribution:

- Shape: It has a characteristic bell-shaped curve.   

- Mean, Median, and Mode: In a normal distribution, the mean (μ), median, and mode are all equal and located at the center of the distribution.   

- Symmetry: The distribution is perfectly symmetric around the mean.   

- 68-95-99.7 Rule (Empirical Rule):

- Approximately 68% of the data falls within 1 standard deviation (σ) of the mean (μ).   
- Approximately 95% of the data falls within 2 standard deviations of the mean.   
- Approximately 99.7% of the data falls within 3 standard deviations of the mean.

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv("D:\\PROJECT_PRACTISE_DIRS\\DSML_PRACTISE_FILES\\DATASETS\\heights.csv")
df.head(3)


mean = df.height.mean()
std = df.height.std()

sns.histplot(df.height, kde=True)

plt.axvline(mean, color='k', linestyle='dashed', linewidth=1)

# one stadard deviation
plt.axvline(mean + std, color='r', linestyle='dashed', linewidth=1)#mu+1sigma
plt.axvline(mean - std, color='r', linestyle='dashed', linewidth=1)#mu-1sigma

# two standard deviation
plt.axvline(mean + 2*std, color='g', linestyle='dashed', linewidth=1)#mu+2sigma
plt.axvline(mean - 2*std, color='g', linestyle='dashed', linewidth=1)#mu-2sigma

# three standard deviation
plt.axvline(mean + 3*std, color='b', linestyle='dashed', linewidth=1)
plt.axvline(mean - 3*std, color='b', linestyle='dashed', linewidth=1)
```

```python
std = round(df.height.std(),2)
mean = round(df.height.mean(),2)

mean, std
```

```python
left1 = round(mean-std,2)
right1 = round(mean+std,2)

left2 = round(mean-2*std,2)
right2 = round(mean+2*std,2)

left3 = round(mean-3*std,2)
right3 = round(mean+3*std,2)

print(left1, right1)
print(left2, right2)
print(left3, right3)
```

```python
print(df[(df.height>left1)&(df.height<right1)].shape)#64% of data lies between 1 std

print(df[(df.height>left2)&(df.height<right2)].shape)#97% of data lies between 2 std

print(df[(df.height>left3)&(df.height<right3)].shape)#99% of data lies between 3 std
```

```python
#Outliers Count if its a normal distribution

print(df[(df.height<left3)|(df.height>right3)].shape)#0.7% of data lies outside
```
