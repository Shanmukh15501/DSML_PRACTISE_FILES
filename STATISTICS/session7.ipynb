# Hypothesis Testing

- Hypothesis testing is a statistical method used to make decisions about a population parameter based on sample data. It involves testing an assumption (called a **hypothesis**) to determine whether there is enough evidence to reject it.

## Key Concepts

- **Null Hypothesis (H₀):** The default assumption that there is no effect or no difference.  
  *Example:* H₀: μ = 100

- **Alternative Hypothesis (H₁ or Ha):** The assumption that contradicts the null hypothesis.  
  *Example:* H₁: μ ≠ 100

- **Significance Level (α):** The probability of rejecting the null hypothesis when it is actually true. Common values are 0.05, 0.01.

- **Test Statistic:** A standardized value calculated from sample data used to decide whether to reject H₀.

- **P-value:** The probability of obtaining results as extreme as the observed results, under the assumption that H₀ is true.

- **Decision Rule:**
  - If *p-value ≤ α*, reject H₀.
  - If *p-value > α*, do not reject H₀

# Type I and Type II Errors

In hypothesis testing, two types of errors can occur when making a decision based on sample data.

## Type I Error (False Positive)

- **Definition:** Rejecting the null hypothesis (H₀) when it is actually true.
- **Symbol:** α (alpha)
- **Interpretation:** Concluding there is an effect or difference when there isn't.
- **Example:** A medical test indicates a patient has a disease when they actually do not.

> **Consequences:** Can lead to unnecessary actions or treatments.

## Type II Error (False Negative)

- **Definition:** Failing to reject the null hypothesis when the alternative hypothesis (H₁) is actually true.
- **Symbol:** β (beta)
- **Interpretation:** Missing a real effect or difference.
- **Example:** A medical test fails to detect a disease in a sick patient.

> **Consequences:** Can result in missed diagnoses or failure to act when needed.

## Comparison Table

| Error Type     | Description                         | Symbol | Reality             | Decision                     |
|----------------|-------------------------------------|--------|----------------------|-------------------------------|
| Type I Error   | False positive                      | α      | H₀ is true           | Reject H₀                    |
| Type II Error  | False negative                      | β      | H₁ is true           | Fail to reject H₀            |

## Minimizing Errors

- Lowering α reduces Type I error but may increase Type II error.
- Increasing sample size helps reduce both types of error.


# Left-tailed, Right-tailed, and Two-tailed Tests

The type of tail test you use depends on the **alternative hypothesis (H₁)**.

## 1. Left-tailed Test

- **Alternative Hypothesis:** H₁: μ < μ₀
- **Critical region is on the left** side of the distribution.
- **Used when testing if the mean is significantly less than a certain value.**

### Example:
- H₀: μ = 100  
- H₁: μ < 100

> Reject H₀ if the test statistic is significantly **less than** expected.

---

## 2. Right-tailed Test

- **Alternative Hypothesis:** H₁: μ > μ₀
- **Critical region is on the right** side of the distribution.
- **Used when testing if the mean is significantly greater than a certain value.**

### Example:
- H₀: μ = 100  
- H₁: μ > 100

> Reject H₀ if the test statistic is significantly **greater than** expected.

---

## 3. Two-tailed Test

- **Alternative Hypothesis:** H₁: μ ≠ μ₀
- **Critical regions are on both sides** of the distribution.
- **Used when testing for any significant difference (either higher or lower).**

### Example:
- H₀: μ = 100  
- H₁: μ ≠ 100

> Reject H₀ if the test statistic is **either much smaller or much larger** than expected.

---

## Choosing the Right Test

|---------------|------------------------|--------------------------------|
| Left-tailed   | H₁: μ < μ₀             | A value is **less than** μ₀    |
| Right-tailed  | H₁: μ > μ₀             | A value is **greater than** μ₀ |
| Two-tailed    | H₁: μ ≠ μ₀             | A value is **different** from μ₀ (either way) |



# Examples of Left-tailed, Right-tailed, and Two-tailed Tests

Below are real-world examples that illustrate when to use each type of hypothesis test.

---

## 1. Left-tailed Test

### Scenario:
A battery manufacturer claims that their batteries last **at least 100 hours**. A consumer group wants to test if the batteries last **less** than 100 hours.

- **Null Hypothesis (H₀):** μ = 100  
- **Alternative Hypothesis (H₁):** μ < 100  
- **Test Type:** Left-tailed

> **Interpretation:** If the test statistic is significantly low (left tail), reject H₀ and conclude the batteries last less than 100 hours.

---

## 2. Right-tailed Test

### Scenario:
A fitness coach claims their new training program increases average vertical jump height to **more than 20 inches**. You want to test if this claim is true.

- **Null Hypothesis (H₀):** μ = 20  
- **Alternative Hypothesis (H₁):** μ > 20  
- **Test Type:** Right-tailed

> **Interpretation:** If the test statistic is significantly high (right tail), reject H₀ and conclude the jump height has increased.

---

## 3. Two-tailed Test

### Scenario:
A pharmaceutical company claims that a drug has **no effect** on blood pressure. A researcher wants to test if the drug **either increases or decreases** blood pressure.

- **Null Hypothesis (H₀):** μ = 0 (no change)  
- **Alternative Hypothesis (H₁):** μ ≠ 0  
- **Test Type:** Two-tailed

> **Interpretation:** If the test statistic is significantly different (either side), reject H₀ and conclude the drug affects blood pressure.

---

## Summary Table

| Test Type     | H₀              | H₁               | When to Use                                  |
|---------------|------------------|------------------|----------------------------------------------|
| Left-tailed   | μ = μ₀           | μ < μ₀           | To check if a value is **less than** μ₀      |
| Right-tailed  | μ = μ₀           | μ > μ₀           | To check if a value is **greater than** μ₀   |
| Two-tailed    | μ = μ₀           | μ ≠ μ₀           | To check if a value is **different from** μ₀ |



```python
from scipy.stats import binom

# Parameters
n = 10          # number of trials
x = 7           # observed heads
p = 0.5         # probability under null hypothesis

# Since we are doing a two-tailed test, we calculate both tails

# Lower tail: P(X <= 3)
lower_tail = binom.cdf(3, n, p)

# Upper tail: P(X >= 7) = 1 - P(X <= 6)
upper_tail = 1 - binom.cdf(6, n, p)

# Two-tailed p-value
p_value = lower_tail + upper_tail

print(f"P-value: {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The coin may not be fair.")
else:
    print("Fail to reject the null hypothesis: No strong evidence the coin is unfair.")
```

## Z-Test
The z-test is a type of hypothesis test used to determine whether there is a significant difference between sample and population means (or between two sample means), when the population variance is known and the sample size is large (typically

## 🧪 Example Scenario
- Suppose the average weight of a certain population is known to be 70 kg with a standard deviation of 8 kg.
- A sample of 50 people has a mean weight of 72.5 kg.
- Is this sample significantly different from the population mean at the 5% significance level?

```python
import numpy as np
from scipy import stats

# Known population parameters
mu = 70           # population mean
sigma = 8         # population standard deviation

# Sample data
sample_mean = 72.5
n = 50

# Calculate z-score
z = (sample_mean - mu) / (sigma / np.sqrt(n))

# Two-tailed test: compute p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Output results
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: no significant difference.")
```

Scenario:

Infosys conducted an Agentic AI training program for all its employees. Before the training, the average employee productivity was 50% with a known standard deviation of 4%. After the training, a random sample of 30 employees showed a mean productivity of 53%.

Question:

Using a z-test, determine whether the Agentic AI training had a statistically significant positive effect on employee productivity. Show the proof using appropriate statistical testing.


### 🔢 Given Data:

- **Population mean (before training):**  $\mu = 50$ 
- **Population standard deviation:** $\sigma = 4$ 
- **Sample size:**  n = 30
- **Sample mean (after training):** $\bar{x} = 53$


```python
import numpy as np
from scipy import stats

# Known population parameters
mu = 50          # population mean before training
sigma = 4        # population std deviation

# Sample data
sample_mean = 53
n = 30

# Z-test statistic
z = (sample_mean - mu) / (sigma / np.sqrt(n))

# One-tailed test (greater than)
p_value = 1 - stats.norm.cdf(z)

# Output
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion at 5% significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The training significantly increased productivity.")
else:
    print("Fail to reject the null hypothesis: No significant improvement detected.")
```

## 📊 Z-Test for Agentic AI Training – Two-Tailed Test

Infosys conducted an Agentic AI training program for its employees. The average employee productivity before the training was known to be 50% with a standard deviation of 4%.

After the training, a random sample of 30 employees had an average productivity of 53%.

---

### 🔢 Given:

- **Population mean:** $\mu = 50$
- **Population standard deviation:** $\sigma = 4$
- **Sample size:** $n = 30$
- **Sample mean:** $\bar{x} = 53$

---

### ❓ Research Question:

Is there **any significant difference** in employee productivity after the training — whether an **increase or decrease**?

We will perform a **two-tailed z-test** at the 5% significance level.

---

### 📌 Hypotheses:

- **Null Hypothesis ($H_0$):** Training has no effect
  $$
  \mu = 50
  $$

- **Alternative Hypothesis ($H_1$):** Training has a significant effect (increase or decrease)
  $$
  \mu \neq 50
  $$

---

### 🧪 Test Statistic:

The formula for the z-test statistic is:

$$
z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}
$$

Substituting the given values:

$$
z = \frac{53 - 50}{\frac{4}{\sqrt{30}}}
$$

$$
z = \frac{3}{\frac{4}{5.477}}
$$

$$
z = \frac{3}{0.730}
$$

$$
z \approx 4.11
$$

---

### 🚦 Critical Values:

For a two-tailed test at a 5% significance level ($\alpha = 0.05$), the critical values are the z-scores that bound the central 95% of the standard normal distribution. These values are $z_{\alpha/2} = z_{0.025}$ and $-z_{\alpha/2} = -z_{0.025}$.

From the standard normal distribution table, the critical values are approximately:

$$
z_{critical} = \pm 1.96
$$

---

### Decision Rule:

Reject the null hypothesis ($H_0$) if the absolute value of the calculated test statistic ($|z|$) is greater than the critical value ($z_{critical}$).

$$
|z| > z_{critical}
$$

---

### Conclusion:

We calculated the test statistic as $z \approx 4.11$. The critical values for a two-tailed test at the 5% significance level are $\pm 1.96$.

Since $|4.11| > 1.96$, we reject the null hypothesis ($H_0$).

**Therefore, there is statistically significant evidence at the 5% significance level to suggest that the Agentic AI training program had a significant effect on employee productivity (either an increase or a decrease).**

```python
import scipy.stats as stats
import math

# Given data
mu_0 = 50       # Population mean
sigma = 4       # Population standard deviation
n = 30          # Sample size
x_bar = 53      # Sample mean
alpha = 0.05    # Significance level

# Step 1: Compute the Z-statistic
z = (x_bar - mu_0) / (sigma / math.sqrt(n))

# Step 2: Critical value for two-tailed test
z_critical = stats.norm.ppf(1 - alpha / 2)

#>>> norm.ppf(1-0.05/2)
#1.959963984540054
#>>> norm.ppf(0.025)
#-1.9599639845400545

# Step 3: P-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))  # Two-tailed

# Print results
print(f"Z-statistic: {z:.4f}")
print(f"Critical value (Z_critical): ±{z_critical:.4f}")
print(f"P-value: {p_value:.4f}")

# Rejection region approach
if abs(z) > z_critical:
    print("Reject H0 using rejection region approach.")
else:
    print("Fail to reject H0 using rejection region approach.")

# P-value approach
if p_value < alpha:
    print("Reject H0 using p-value approach.")
else:
    print("Fail to reject H0 using p-value approach.")
```

🚫 **Why Z-Test is Not Recommended for Small Sample Sizes (e.g., n=5)**
- Z-test assumes a large sample size (typically 𝑛 ≥ 30 n≥30) or that the population is normally distributed with known variance.
- With small samples, the Central Limit Theorem doesn’t apply well, so the sampling distribution of the mean may not be normal.
- Therefore, the z-test becomes unreliable unless you're absolutely sure the population is perfectly normal.

# p-value

- The "p-test" refers to using the p-value in hypothesis testing. The p-value is the probability of observing your data (or more extreme data) if the null hypothesis were true.
- It is the measure of strength against null hypothesis
- You compare the p-value to a pre-set significance level (α).
- If p≤α: Reject the null hypothesis (statistically significant).
- If p>α: Fail to reject the null hypothesis (not statistically significant).

## Z-Test Overview

**Use the Z-test when:**

* Population standard deviation $\sigma$ is known
* Sample size is large ($n \ge 30$) or population is normally distributed

**Z-test formula:**
$$z = \frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}}$$

Where:

* $\bar{x}$: sample mean
* $\mu_0$: population mean under the null hypothesis
* $\sigma$: population standard deviation
* $n$: sample size

### ✅ Example 1: Right-tailed Z-test

**Claim:** A company claims their bulbs last more than 1000 hours on average.

$\mu_0 = 1000$
$\sigma = 40$
Sample: $\bar{x} = 1015$, $n = 36$
$\alpha = 0.05$

**Step 1: Hypotheses**
$H_0: \mu = 1000$
$H_a: \mu > 1000$ (Right-tailed test)

**Step 2: Compute z-statistic**
$$z = \frac{1015 - 1000}{40 / \sqrt{36}} = \frac{15}{40 / 6} = \frac{15}{6.67} \approx 2.25$$

**Step 3: Find p-value**
Using a Z-table or calculator:

For $z = 2.25$, area to the right (since it’s right-tailed):
$p = P(Z > 2.25) \approx 0.0122$

**Step 4: Compare with $\alpha = 0.05$**
$p = 0.0122 < 0.05 \implies$ Reject $H_0$

✅ **Conclusion:** There is sufficient evidence to support the claim that the bulbs last more than 1000 hours.

### ✅ Example 2: Two-tailed Z-test

**Claim:** A machine fills bottles with exactly 500 ml. We suspect it might be off.

$\mu_0 = 500$
$\sigma = 5$
Sample: $\bar{x} = 498.5$, $n = 64$

**Step 1: Hypotheses**
$H_0: \mu = 500$
$H_a: \mu \ne 500$ (Two-tailed test)

**Step 2: Compute z-statistic**
$$z = \frac{498.5 - 500}{5 / \sqrt{64}} = \frac{-1.5}{5 / 8} = \frac{-1.5}{0.625} = -2.4$$

**Step 3: Find p-value**
$z = -2.4$
Two-tailed: Find area in both tails

$P(Z < -2.4) \approx 0.0082 \implies \text{Two-tailed } p = 2 \times 0.0082 = 0.0164$

**Step 4: Compare with $\alpha = 0.05$**
$\alpha = 0.05$
$p = 0.0164 < 0.05 \implies$ Reject $H_0$

✅ **Conclusion:** There is significant evidence that the mean is different from 500 ml.

## Example 4
- Example (Right-Tailed Z-Test) A company claims that their bulbs last more than 1000 hours.
- Population mean under $H_0$: $\mu_0 = 1000$
- Sample mean: $\bar{x} = 1015$
- Population standard deviation: $\sigma = 40$
- Sample size: $n = 36$
- Significance level: $\alpha = 0.05$

```python
import scipy.stats as stats
import math

# Given data
mu_0 = 1000        # Hypothesized mean
x_bar = 1015       # Sample mean
sigma = 40         # Population standard deviation
n = 36             # Sample size
alpha = 0.05       # Significance level

# Step 1: Compute z-statistic
z = (x_bar - mu_0) / (sigma / math.sqrt(n))

# Step 2: Critical value for right-tailed test
z_critical = stats.norm.ppf(1 - alpha)

# Step 3: P-value
p_value = 1 - stats.norm.cdf(z)

# Print results
print(f"Z-statistic: {z:.4f}")
print(f"Critical value (Z_critical): {z_critical:.4f}")
print(f"P-value: {p_value:.4f}")

# Rejection region approach
if z > z_critical:
    print("Reject H0 using rejection region approach.")
else:
    print("Fail to reject H0 using rejection region approach.")

# P-value approach
if p_value < alpha:
    print("Reject H0 using p-value approach.")
else:
    print("Fail to reject H0 using p-value approach.")
```
